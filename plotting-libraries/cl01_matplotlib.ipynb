```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
    y
```

```python
plt.plot(x,y)
plt.xlabel("this is x-axis")
plt.ylabel("y-axis")
plt.title("sin func")
plt.show()
#represents the sin function data plotting
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
size=100*np.random.rand(50)
plt.scatter(x,y,c=colors,s=size,alpha=0.9)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("random data")
```
