```python
#here we will have to remove all the other symbols than the numeric characters 
df['Installs'].unique()

```

```python
#the same thing with the price feature dollar sign and all 
df['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']

for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')

```

# now as working on the last updated feature so 
take out the day month and year and make separate rows for them in the df_copy

```python
df_copy['Last Updated']
#just watch at its format 
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])

#separating day month and year 
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year


```

```python
df_copy.info()
```

# EDA PART STARTS FROM HERE 

```python
df_copy[df_copy.duplicated('App')]
```

```python
df_copy[df_copy.duplicated('App')].shape
#these many values are duplicates under the data
```

```python
#deleting all the duplicates using a function 
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy[df_copy.duplicated('App')].shape
#here you can see all duplicates got deleted 
```

```python
# check out for numerical and categorical data

numerical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']


#print the columns
print("we have {} numerical features :{}".format(len(numerical_features),numerical_features))
print("\n we have {} categorical features :{}".format(len(categorical_features),categorical_features))
```

# visualising the data 

```python
plt.figure(figsize=(15,15))

plt.suptitle("univariate analysis of numerical features ",fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numerical_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numerical_features[i]],shade=True,color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

# insights 

## Rating and Year,last updated   is left skewed while Reviews ,size,installs ,price are right skewed 

```python
#plotting the same graph for the categorical variables 

plt.figure(figsize=(20,15))

plt.suptitle("univariate analysis of categorical  features ",fontsize=20,fontweight='bold',alpha=0.8,y=1.)
category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```
