# decision tree classifier implementation 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
            print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
#choosing dependent and independent features 

X=df.iloc[:,:-1]
y=dataset.target

X,y
```

```python
#do the train test split 
#note dont do standarisation for the train test split because the values will get decreased to smaller unit  

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

treeClassifier=DecisionTreeClassifier()
```

```python
treeClassifier.fit(X_train,y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# now we will plot the tree 
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
#post pruning technique with max_depth=2s
#doing only 2 level of split

classifier1=DecisionTreeClassifier(max_depth=2)

classifier1.fit(X_train,y_train)
```

```python
tree.plot_tree(classifier1,filled=True)
```

```python
#do the prediction 
y_pred=classifier1.predict(X_test)
```

```python
y_pred
```

```python
#check accuracy and classification report
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
#classification report
print(classification_report(y_pred,y_test))
```

# decision tree prepruning(for huge data) and hyper parameter tuning 
|

```python
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']     
    
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier=DecisionTreeClassifier()
```

```python
clf=GridSearchCV(treeclassifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
#train the model with data
clf.fit(X_train,y_train)
```

```python
#predict and match with the previously predicted data
y_pred=clf.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```
