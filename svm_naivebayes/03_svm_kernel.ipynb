# svm kernel practical implementation

```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.linspace(-5.0,5.0,100)
y=np.sqrt(10**2-x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x1=np.linspace(-5.0,5.0,100)
y1=np.sqrt(5**2-x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.figure(figsize=(15,10))
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
import pandas as pd
df1=pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2=pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df= pd.concat([df1,df2],ignore_index=True)
df.head()
```

```python
df.tail()
```

# apply the polynomial kernel 



# based on the formula find the components for the polynomial kernel 

```python
df['X1_Square']=df['X1']**2
df['X2_Square']=df['X2']**2
df['X1 * X2']=df['X1']*df['X2']
df.head()
```

# dependent feature and independent feature 

```python
X=df[['X1_Square','X2_Square','X1 * X2']]
y=df['Y']
```

# train test split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

# Radial Basis Function Kernel



```python
df.head()
```

```python
## Indpeendent Features
X=df.iloc[:,0:2]
y=df.Y
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.25, 
                                                    random_state = 0)
```

```python
classifier = SVC(kernel="rbf")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
classifier = SVC(kernel="poly")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

# sigmoid kernel


```python
classifier = SVC(kernel="sigmoid")
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
